>Using the data below, answer the following questions:

>1) Which entities (top 5) had the largest population density in 2020?
>2) Which entities have more water area than land area?
>3) Which entities increased in population the most in the last 10 years?
>4) What state bird accounts for the largest population as of 2020? Land area?
>5) How many entities' largest city is their capital city?
>6) Which city has the largest percent drop from their largest city to their 5th largest? 100*(1st largest - 5th largest)/(1st largest)

```python
import pandas as pd
```

```python
facts = pd.read_csv('../data/state_facts.tsv',delimiter="\t")
facts.head(5)
```

```python
#1) Which entities (top 5) had the largest population density in 2020?
facts['Population_density']=facts['Pop_2020']/facts['Area_land']
popDens = facts.sort_values(by='Population_density',ascending=False)
facts.head(5)
```

```python
#2) Which entities have more water area than land area?
waterLand = facts[facts['Area_land'] < facts['Area_water']]
waterLand
```

```python
#3) Which entities increased in population the most in the last 10 years?
facts['Population_inc'] = facts['Pop_2020'] - facts['Pop_2010']
increPop = facts.sort_values(by='Population_inc',ascending=False)
increPop.head(5)
```

```python
#4) What state bird accounts for the largest population as of 2020? Land area?
birds = facts.sort_values(by='Pop_2020',ascending=False)
print(birds['State_bird'].head(1))
print("")

landArea = facts.sort_values(by='Area_land',ascending=False)
print(landArea['State_bird'].head(1))
```

```python
#5) How many entities' largest city is their capital city?
count = 0
facts['Capital=LargestCity'] = facts['Capital'] == facts['City_1']
for i in range(56):
    if facts['Capital=LargestCity'][i]==True:
        count += 1
count       
```

```python
#6) Which city has the largest percent drop from their largest city to their 5th largest? 100*(1st largest - 5th largest)/(1st largest)
facts['LargestPercentDrop'] = (facts['city_1_pop'] - facts['city_5_pop']) / facts['city_1_pop'] * 100
largest = facts.sort_values(by='LargestPercentDrop', ascending=False)
largest.head(1)
```
